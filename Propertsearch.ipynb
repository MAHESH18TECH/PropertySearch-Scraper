```python
import pandas as pd
import random
import time
from parsel import Selector
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys

chrome_options = Options()

PROXY = "134.238.252.143:8080" #  HOST:PORT

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument(f'--proxy-server={PROXY}')

driver = webdriver.Chrome(executable_path=r'chromedriver', options=chrome_options)

driver.get('https://propertysearch.altusgroup.com/Property/Search/All/All/For-Sale-and-To-Let/All/All/All/All/recently-added-first/All')

time.sleep(random.uniform(2,3))

load_more = driver.find_element(By.XPATH,'//button[@class="load-more-button d-inline-block title3"]')
for i in range(50):
    if i < 50:
        try:
            load_more.click()
            sleep(5)
        except:
            pass
        
print('Page Loaded Succesfully')
lnks = []
urls = driver.find_elements(By.XPATH, "//div[@class='property-card d-block w-100']/a[@href]")
for url in urls:
    all_links = url.get_attribute("href")
    lnks.append(all_links)
all_data = []   
for links in lnks:
    driver.get(links)
    

    property_name = driver.find_element(By.XPATH,"/html/body/main/div/div[2]/div/div/div[2]/h2")
    name = property_name.text
    property_area = driver.find_element(By.XPATH,"/html/body/main/div/div[2]/div/div/div[2]/p[1]")
    area = property_area.text
    key_features = driver.find_elements(By.XPATH,"//p[@class = 'd-block key-feature-item title5']")
    for y in key_features:
        features = y.text
    agent_name = driver.find_element(By.XPATH,"/html/body/main/div/div[3]/div/div/div[2]/div/div[1]/div[2]/p")
    agent = agent_name.text
    agent_email = driver.find_element(By.XPATH,"/html/body/main/div/div[3]/div/div/div[2]/div/div[1]/div[2]/a[1]")
    agnt_eml = agent_email.get_attribute("href")
    agent_number = driver.find_element(By.XPATH,"/html/body/main/div/div[3]/div/div/div[2]/div/div[1]/div[2]/a[2]")
    agnt_nmbr = agent_number.get_attribute("href")

    data = {
        'Property Name':name,
        'Area':area,
        'Key Features':features,
        'Agent Name': agent,
        'Agent Email': agnt_eml,
        'Agent Number': agnt_nmbr
            }

    all_data.append(data)
    
    
df = pd.DataFrame(all_data)
df.to_excel("Property_Details.xlsx")
```
